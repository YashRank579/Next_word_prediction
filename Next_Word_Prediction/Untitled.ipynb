```python
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# 1. Load the model
loaded_model = load_model('next_word_model.keras')

# 2. Load the tokenizer
with open('tokenizer.pickle', 'rb') as handle:
    loaded_tokenizer = pickle.load(handle)

# 3. Define the prediction function again (using loaded model and tokenizer)
def predict_with_loaded_model(seed_text, next_words):
    for _ in range(next_words):
        token_list = loaded_tokenizer.texts_to_sequences([seed_text])[0]
        # Note: We use the max_sequence_len from training (57 - 1 = 56)
        token_list = pad_sequences([token_list], maxlen=56, padding='pre')
        
        predicted = np.argmax(loaded_model.predict(token_list, verbose=0), axis=-1)
        
        output_word = ""
        for word, index in loaded_tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        
        seed_text += " " + output_word
    return seed_text

# Test the loaded model
print(predict_with_loaded_model("what is the fee", 5))
```
